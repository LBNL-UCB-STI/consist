## Model Overview

We simulate morning commute mode choice in a stylized 5-zone linear city:

```
Zone 1 ←→ Zone 2 ←→ Zone 3 ←→ Zone 4 ←→ Zone 5
(Low)     (Med)     (High/CBD)  (Med)     (Low)
```

**Feedback loop:**
1. **Logsums**: Compute accessibility from current travel times
2. **Trip Distribution**: Workers choose job locations based on accessibility
3. **Mode Choice**: Each commuter chooses car, transit, or walk
4. **Assignment**: Car trips create congestion
5. **Skim Update**: Congestion slows travel times → back to step 1

The model iterates until mode shares stabilize.


## Workflow Structure (Iteration Loop)

This is the conceptual loop we'll run in each scenario:

```
┌─────────────┐
│    init     │ (iteration 0 only)
└──────┬──────┘
       │ skims, zones, population
       ▼
┌─────────────┐
│   logsums   │◄─── skims, zones
└──────┬──────┘
       │ logsums
       ▼
┌─────────────┐
│ trip_dist   │◄─── logsums, zones, population
└──────┬──────┘
       │ trips
       ▼
┌─────────────┐
│ utilities   │◄─── trips, skims, zones
└──────┬──────┘
       │ utilities
       ▼
┌─────────────┐
│ mode_choice │◄─── utilities, (prev trips_with_modes)
└──────┬──────┘
       │ trips_with_modes
       ▼
┌─────────────┐
│ assignment  │◄─── trips_with_modes
└──────┬──────┘
       │ volumes
       ▼
┌─────────────┐
│ traffic_sim │◄─── volumes, skims
└──────┬──────┘
       │ updated skims ──► next iteration
       ▼
```

For `run(...)` vs `trace(...)`, see the [Usage Guide](../docs/usage-guide.md#when-does-code-execute-understanding-scrun-vs-sctrace).

## Model Configuration

The model has four parameter groups:

| Parameter Group           | Controls                                                     |
|---------------------------|--------------------------------------------------------------|
| `ZoneParams`              | City layout: population, jobs, parking costs, transit access |
| `ModeChoiceParams`        | Utility coefficients, speeds, costs                          |
| `DestinationChoiceParams` | How accessibility affects job location choice                |
| `AssignmentParams`        | BPR congestion function parameters                           |

In typical TDM practice, these parameters live in scattered config files, spreadsheets,
and hardcoded values. When you return to a run months later, reconstructing exactly
what was used is painful. Consist captures all of this in a queryable format.

**ZoneParams** defines the city structure. The central zone (Zone 3) has the most jobs and highest
parking costs, mimicking a downtown CBD. Outer zones are residential with cheaper or free parking.

**ModeChoiceParams** sets coefficients for a standard multinomial logit. `beta_cost` and
`beta_time` control sensitivity to cost/time; ASCs capture baseline preferences.

**DestinationChoiceParams** governs how workers choose job locations based on accessibility
(logsum) and employment size.

**AssignmentParams** controls the BPR (Bureau of Public Roads) function that converts volumes into
congested travel times.


````python
zone_params = ZoneParams()
mode_params = ModeChoiceParams()
dest_params = DestinationChoiceParams()
assignment_params = AssignmentParams()

DEFAULT_SEED = 0

base_config = TravelDemandScenarioConfig(
    n_iterations=10,
    seed=DEFAULT_SEED,
    zone_params=zone_params,
    mode_params=mode_params,
    dest_params=dest_params,
    assignment_params=assignment_params,
)

SCENARIO_NAME = "travel_demand_demo"

SKIM_PERTURBATION = 0.1
DESTINATION_UPDATE_SHARE = 0.2
````

## Workflow Functions

The next cell defines the model steps and the scenario runner. It is long, but it mirrors a real TDM pipeline: initialize inputs, compute logsums, distribute trips, choose modes, assign volumes, update congested skims, and write scenario summaries.

It also intentionally shows a few Consist integration styles side by side. That is useful for a demonstration notebook because different wrappers need different levels of control, but in a production workflow you would usually pick one house style and apply it consistently.

The main patterns are:

- **Context-injected steps**: `initialize_scenario(..., _consist_ctx)` and `summarize_results_step(..., _consist_ctx)` request a `RunContext` with `ExecutionOptions(inject_context=True)`. Use this when the step needs run-managed output paths such as `_consist_ctx.output_path(...)` or wants to log path-based outputs explicitly with `_consist_ctx.log_output(...)`.
- **Global logging helpers**: several steps call `consist.log_dataframe(...)` directly. This is concise inside an active run and works well when a function creates an in-memory DataFrame and wants Consist to persist it as an output artifact.
- **Return-value functions**: functions imported from `examples/src/travel_demand_functions.py`, such as `compute_mode_utilities`, `apply_mode_choice`, and `apply_congestion`, are passed directly to `scenario.run(...)`. They do not call Consist themselves; `outputs=[...]` tells Consist how to log their returned values.
- **Loaded artifact binding**: most iterative steps use `ExecutionOptions(input_binding="loaded")`. That tells Consist to resolve input artifact references like `consist.ref(logsums_result, key="logsums")`, load the artifact bytes, and pass DataFrames or datasets into function arguments by name.
- **Runtime-only kwargs**: values such as `mode_params`, `dest_params`, and seeds are passed through `runtime_kwargs={...}` when the Python function needs the original object at execution time. The separate `config=...` payload is the serialized identity that Consist hashes, stores, diffs, and indexes.
- **No input binding for summary bookkeeping**: the final summary step uses `input_binding="none"` because the summary data already lives in local dictionaries accumulated by the loop. It still declares the final trips artifact as an input for provenance, but it does not need Consist to bind that artifact into the function call.

The code is therefore a small catalog of integration options. For a real model wrapper, prefer the simplest convention that fits the tool: direct return-value logging for pure Python functions, context-injected paths for file-writing steps, and requested/path-bound staging when an external executable needs fixed filenames.

````python
from consist import CacheOptions, ExecutionOptions


# Context-injected steps use _consist_ctx for run-managed output paths and
# explicit path-based artifact logging. This is useful for file-backed outputs
# such as Zarr stores, directories, or external-model workspaces.
def initialize_scenario(zone_params, mode_params, skim_perturbation, _consist_ctx):
    zones = generate_zones(zone_params)
    population = generate_population(zones)
    distances = generate_distances(zone_params)

    skims = create_skims_dataset(zones, distances, mode_params)
    skims["time_car_mins"] *= skim_perturbation

    # Ask Consist for a stable output location inside this run's artifact
    # directory. This is just a Path; the model code still writes the Zarr
    # store itself, then we log that path under the semantic key "skims".
    input_skims_path = _consist_ctx.output_path("skims_init", ext="zarr")

    save_skims(skims, input_skims_path)

    _consist_ctx.log_output(input_skims_path, key="skims")
    consist.log_dataframe(
        population,
        key="persons",
        direction="output",
    )
    consist.log_dataframe(zones, key="zones", direction="output")


# This step logs an in-memory DataFrame with the concise active-run helper.
# It does not need a RunContext because Consist chooses the storage path.
def compute_logsums_step(skims, zones, mode_params):
    logsums = compute_od_logsums(
        skims,
        zones,
        mode_params,
    )
    consist.log_dataframe(
        logsums.to_dataframe().reset_index(),
        key="logsums",
    )


# Another direct logging style: load inputs as DataFrames, compute a DataFrame,
# and let consist.log_dataframe(...) persist it under a stable artifact key.
def distribute_trips_step(
    zones, persons, logsums, dest_params, seed, update_share, trips=None
):
    population = persons
    prev_trips = trips
    logsums_df = logsums
    if "origin" not in logsums_df.columns or "destination" not in logsums_df.columns:
        logsums_df = logsums_df.reset_index()
    logsums_da = logsums_df.set_index(["origin", "destination"])["logsum"].to_xarray()
    trips_df = distribute_trips(
        population,
        zones,
        logsums_da,
        dest_params,
        seed=seed,
        prev_trips=prev_trips,
        update_share=update_share,
    )
    consist.log_dataframe(trips_df, key="trips")


# Summary outputs use both styles: _consist_ctx chooses stable managed file
# paths, then consist.log_dataframe(..., path=...) writes named CSV artifacts.
def summarize_results_step(mode_shares, summaries, pmt_totals, _consist_ctx):
    mode_shares_df = (
        pd.DataFrame.from_dict(mode_shares, orient="index")
        .sort_index()
        .rename_axis("iteration")
        .reset_index()
    )
    summaries_df = (
        pd.DataFrame.from_dict(summaries, orient="index")
        .sort_values("iteration")
        .reset_index(drop=True)
    )
    pmt_totals_df = (
        pd.DataFrame.from_dict(pmt_totals, orient="index")
        .sort_index()
        .rename_axis("iteration")
        .reset_index()
    )

    mode_shares_path = _consist_ctx.output_path("summary/mode_shares", ext="csv")
    summaries_path = _consist_ctx.output_path("summary/iteration_summaries", ext="csv")
    pmt_totals_path = _consist_ctx.output_path("summary/pmt_totals", ext="csv")

    consist.log_dataframe(
        mode_shares_df,
        key="mode_shares",
        path=mode_shares_path,
    )
    consist.log_dataframe(
        summaries_df,
        key="iteration_summaries",
        path=summaries_path,
    )
    consist.log_dataframe(
        pmt_totals_df,
        key="pmt_totals",
        path=pmt_totals_path,
    )


# The scenario runner wires the steps together. The examples below intentionally
# show several binding/logging styles; in production, pick the smallest style
# that fits your model wrapper and use it consistently.
def run_scenario(scenario_config, scenario_run_id):
    mode_shares = {}
    summaries = {}
    pmt_totals = {}

    zone_params = scenario_config.zone_params
    mode_params = scenario_config.mode_params
    dest_params = scenario_config.dest_params
    assignment_params = scenario_config.assignment_params
    seed = scenario_config.seed

    with consist.scenario(
        scenario_run_id,
        tracker=tracker,
        config={
            **asdict(scenario_config),
            "scenario_name": SCENARIO_NAME,
        },
        facet_from=[
            "n_iterations",
            "seed",
            "zone_params",
            "mode_params",
            "dest_params",
            "assignment_params",
            "scenario_name",
        ],
        tags=["examples", "simulation", "travel_demand"],
    ) as scenario:
        scenario.require_outputs(
            "skims",
            "zones",
            "persons",
            "logsums",
            "trips",
            "utilities",
            "trips_with_modes",
            "volumes",
            "mode_shares",
            "iteration_summaries",
            "pmt_totals",
        )

        cache_validation = "lazy"  # Skip output checks on cache hits for speed; use 'eager' to validate files.
        cache_hydration = (
            "inputs-missing"  # Copy cached inputs into the new run_dir on cache misses.
        )

        # Phase 1: Build baseline scenario inputs and seed artifacts.
        init_result = scenario.run(
            name="init",
            fn=initialize_scenario,
            config=asdict(mode_params)
            | asdict(zone_params)
            | {"skim_perturbation": SKIM_PERTURBATION},
            facet_from=["skim_perturbation"],
            outputs=["skims", "persons", "zones"],
            execution_options=ExecutionOptions(
                # inject_context=True supplies the _consist_ctx argument above.
                inject_context=True,
                runtime_kwargs={
                    "zone_params": zone_params,
                    "mode_params": mode_params,
                    "skim_perturbation": SKIM_PERTURBATION,
                },
            ),
        )

        init_refs = consist.refs(init_result, "zones", "persons")
        zones_ref = init_refs["zones"]
        persons_ref = init_refs["persons"]
        skims_result = init_result
        previous_trip_distribution_result = None
        latest_mode_choice_result = None

        # Phase 2: Run iterative demand/assignment feedback.
        print(f"Running {scenario_config.n_iterations} iterations...")
        for i in range(scenario_config.n_iterations):
            # 2a: Compute logsums from current network conditions.
            logsums_result = scenario.run(
                name="logsums",
                fn=compute_logsums_step,
                run_id=f"{scenario.run_id}_logsums_{i}",
                config=asdict(mode_params),
                inputs={
                    "skims": consist.ref(skims_result, key="skims"),
                    "zones": zones_ref,
                },
                facet_from=[
                    "beta_cost",
                    "beta_time",
                    "asc_walk",
                    "asc_transit",
                    "fuel_cost_per_mile",
                    "transit_fare",
                ],
                cache_options=CacheOptions(
                    validate_cached_outputs=cache_validation,
                    cache_hydration=cache_hydration,
                ),
                execution_options=ExecutionOptions(
                    # input_binding="loaded" hydrates artifact refs into
                    # function arguments such as skims, zones, and logsums.
                    input_binding="loaded",
                    runtime_kwargs={"mode_params": mode_params},
                ),
                outputs=["logsums"],
                iteration=i,
            )

            trip_inputs = {
                "zones": zones_ref,
                "persons": persons_ref,
                "logsums": consist.ref(logsums_result, key="logsums"),
            }
            if previous_trip_distribution_result is not None:
                trip_inputs["trips"] = consist.ref(
                    previous_trip_distribution_result, key="trips"
                )

            # 2b: Update destination choice (trip distribution).
            trip_distribution_result = scenario.run(
                name="trip_distribution",
                fn=distribute_trips_step,
                run_id=f"{scenario.run_id}_trip_distribution_{i}",
                config={
                    **asdict(dest_params),
                    "update_share": DESTINATION_UPDATE_SHARE,
                },
                inputs=trip_inputs,
                facet_from=["beta_size", "beta_access", "update_share"],
                outputs=["trips"],
                execution_options=ExecutionOptions(
                    input_binding="loaded",
                    runtime_kwargs={
                        "dest_params": dest_params,
                        "seed": seed + i,
                        "update_share": DESTINATION_UPDATE_SHARE,
                    },
                ),
                iteration=i,
            )

            # 2c: Recompute mode utilities and simulate mode choices.
            # These imported functions return DataFrames directly; outputs=[...]
            # tells Consist which artifact key to use for the return value.
            utilities_result = scenario.run(
                name="calculate_utilities",
                fn=compute_mode_utilities,
                run_id=f"{scenario.run_id}_utilities_{i}",
                config=asdict(mode_params),
                inputs={
                    "trips": consist.ref(trip_distribution_result, key="trips"),
                    "skims": consist.ref(skims_result, key="skims"),
                    "zones": zones_ref,
                },
                facet_from=[
                    "beta_cost",
                    "beta_time",
                    "asc_walk",
                    "asc_transit",
                ],
                cache_options=CacheOptions(
                    validate_cached_outputs=cache_validation,
                    cache_hydration=cache_hydration,
                ),
                execution_options=ExecutionOptions(
                    input_binding="loaded",
                    runtime_kwargs={"mode_params": mode_params},
                ),
                outputs=["utilities"],
                iteration=i,
            )

            mode_choice_result = scenario.run(
                name="mode_choice",
                fn=apply_mode_choice,
                run_id=f"{scenario.run_id}_mode_choice_{i}",
                config={"seed": seed},
                inputs={"utilities_df": consist.ref(utilities_result, key="utilities")},
                execution_options=ExecutionOptions(
                    input_binding="loaded",
                    runtime_kwargs={"seed": seed + i},
                ),
                outputs=["trips_with_modes"],
                iteration=i,
            )

            # 2d: Assign trips to network and update congested skims.
            assignment_result = scenario.run(
                name="assignment",
                fn=compute_od_volumes,
                run_id=f"{scenario.run_id}_assignment_{i}",
                inputs={
                    "trips": consist.ref(mode_choice_result, key="trips_with_modes")
                },
                execution_options=ExecutionOptions(input_binding="loaded"),
                outputs=["volumes"],
                iteration=i,
            )

            skims_result = scenario.run(
                name="traffic_simulation",
                fn=apply_congestion,
                run_id=f"{scenario.run_id}_traffic_simulation_{i}",
                config=asdict(assignment_params),
                inputs={
                    "volumes": consist.ref(assignment_result, key="volumes"),
                    "skims": consist.ref(skims_result, key="skims"),
                },
                facet_from=["bpr_alpha", "bpr_beta", "base_capacity"],
                execution_options=ExecutionOptions(
                    input_binding="loaded",
                    runtime_kwargs={"assignment_params": assignment_params},
                ),
                outputs=["skims"],
                iteration=i,
            )

            # 2e: Capture per-iteration summary metrics for reporting.
            trips_with_modes = consist.load_df(
                consist.ref(mode_choice_result, key="trips_with_modes"), tracker=tracker
            )
            shares = compute_mode_shares(trips_with_modes)
            mode_shares[i] = pd.Series(shares)
            distance_by_mode = trips_with_modes.groupby("mode")["distance_miles"].sum()
            pmt_totals[i] = distance_by_mode.sort_index()
            summaries[i] = summarize_iteration(i, trips_with_modes, shares, 0, False)

            previous_trip_distribution_result = trip_distribution_result
            latest_mode_choice_result = mode_choice_result

        if latest_mode_choice_result is None:
            raise RuntimeError("No iterations were run; cannot summarize mode shares.")

        # Phase 3: Persist scenario-level summary artifacts.
        summary_result = scenario.run(
            name="summaries",
            fn=summarize_results_step,
            inputs={
                "trips_with_modes": consist.ref(
                    latest_mode_choice_result, key="trips_with_modes"
                )
            },
            outputs=["mode_shares", "iteration_summaries", "pmt_totals"],
            cache_options=CacheOptions(
                validate_cached_outputs=cache_validation,
                cache_hydration=cache_hydration,
            ),
            execution_options=ExecutionOptions(
                # The summary dictionaries are runtime state, not artifact inputs
                # to bind into the function call, so binding is disabled here.
                input_binding="none",
                inject_context=True,
                runtime_kwargs={
                    "mode_shares": mode_shares,
                    "summaries": summaries,
                    "pmt_totals": pmt_totals,
                },
            ),
        )

        mode_shares_df = consist.load_df(summary_result.outputs["mode_shares"])
        summaries_df = consist.load_df(summary_result.outputs["iteration_summaries"])
        pmt_totals_df = consist.load_df(summary_result.outputs["pmt_totals"])

    return {
        "scenario_run_id": scenario_run_id,
        "mode_shares_df": mode_shares_df,
        "summaries_df": summaries_df,
        "pmt_totals_df": pmt_totals_df,
    }
````

## Scenario Comparison: Higher Parking Costs

Now run a policy scenario with higher CBD parking costs and compare to the baseline. This is a
common TDM workflow: a scenario change followed by summary comparisons.


````python
high_parking_config = replace(
    base_config,
    zone_params=ZoneParams(parking_costs=(0.0, 8.0, 25.0, 8.0, 0.0)),
)
high_parking_run_id = f"{SCENARIO_NAME}_{SESSION_ID}_high_parking"
high_parking_results = run_scenario(high_parking_config, high_parking_run_id)

high_parking_step_runs = tracker.find_runs(parent_id=high_parking_run_id)
high_parking_cache_hits = [
    run for run in high_parking_step_runs if (run.meta or {}).get("cache_hit")
]
print(
    f"High parking cache hits: {len(high_parking_cache_hits)}/{len(high_parking_step_runs)} "
    "steps served from cache."
)

base_shares = base_results["mode_shares_df"].set_index("iteration").loc[final_iteration]
hp_shares = (
    high_parking_results["mode_shares_df"].set_index("iteration").loc[final_iteration]
)
mode_share_comparison = pd.DataFrame(
    {
        "base": base_shares,
        "high_parking": hp_shares,
    }
)
mode_share_comparison["delta"] = (
    mode_share_comparison["high_parking"] - mode_share_comparison["base"]
)

base_pmt = base_results["pmt_totals_df"].set_index("iteration").loc[final_iteration]
hp_pmt = (
    high_parking_results["pmt_totals_df"].set_index("iteration").loc[final_iteration]
)
pmt_comparison = pd.DataFrame(
    {
        "base": base_pmt,
        "high_parking": hp_pmt,
    }
)
pmt_comparison["delta"] = pmt_comparison["high_parking"] - pmt_comparison["base"]

summary_cols = ["vmt", "total_trips", "car_share", "transit_share", "walk_share"]
base_summary = (
    base_results["summaries_df"]
    .set_index("iteration")
    .loc[final_iteration, summary_cols]
)
hp_summary = (
    high_parking_results["summaries_df"]
    .set_index("iteration")
    .loc[final_iteration, summary_cols]
)
summary_comparison = pd.DataFrame(
    {
        "base": base_summary,
        "high_parking": hp_summary,
    }
)
summary_comparison["delta"] = (
    summary_comparison["high_parking"] - summary_comparison["base"]
)

display(mode_share_comparison)
display(pmt_comparison)
display(summary_comparison)
````

````python
car_delta_pp = mode_share_comparison.loc["car", "delta"] * 100
transit_delta_pp = mode_share_comparison.loc["transit", "delta"] * 100
walk_delta_pp = mode_share_comparison.loc["walk", "delta"] * 100
vmt_delta = summary_comparison.loc["vmt", "delta"]
vmt_base = summary_comparison.loc["vmt", "base"]
vmt_pct = (vmt_delta / vmt_base) * 100 if vmt_base else float("nan")

fig, ax = plt.subplots(figsize=(5, 3))
mode_share_comparison[["base", "high_parking"]].plot(kind="bar", ax=ax)
ax.set_ylabel("Mode Share")
ax.set_title("Mode Shift: High Parking Scenario")
ax.legend(["Baseline", "High Parking"])
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

print(
    f"High parking shifts {abs(car_delta_pp):.1f} pp out of car, "
    f"with transit up {abs(transit_delta_pp):.1f} pp and walk {abs(walk_delta_pp):.1f} pp."
)
print(f"VMT changes by {vmt_pct:.1f}% (Δ={vmt_delta:.2f}).")
````

## Provenance Queries

### Config diff

`tracker.diff_runs()` compares flattened config facets between two scenario runs and returns
only the keys that differ. A parking-cost-only change should show exactly one changed key.

````python
diff = tracker.diff_runs(SCENARIO_RUN_ID, high_parking_results["scenario_run_id"])

diff_rows = []
for key in sorted(diff["changes"]):
    change = diff["changes"][key]
    if change["status"] == "same":
        continue
    diff_rows.append(
        {
            "key": key,
            "baseline": change.get("left"),
            "policy": change.get("right"),
        }
    )

diff_df = pd.DataFrame(diff_rows)
if diff_df.empty:
    print("No differences found between scenario facets.")
diff_df
````

## RunSet: Structured Multi-Run Analysis

`RunSet` provides a composable API for working with collections of runs — filtering, grouping,
aligning across scenarios, and applying functions to matched pairs.

Here we align the baseline and high-parking `summaries` runs 1:1 and compute outcome deltas
(mode shares, PMT, VMT) across scenarios in a single `apply()` call.

````python
all_runs = tracker.run_set(label="all-runs", status="completed")

baseline_runs = all_runs.filter(parent_id=SCENARIO_RUN_ID)
policy_runs = all_runs.filter(parent_id=high_parking_results["scenario_run_id"])

# Align the single summaries run from each scenario for pairwise comparison.
summary_pair = (
    baseline_runs.filter(model="summaries")
    .latest()
    .align(policy_runs.filter(model="summaries").latest(), on="model")
)


def compare_outcomes(left_run, right_run, _model_key):
    """Load summary artifacts from a matched run pair and return per-metric deltas."""
    left_out = tracker.get_run_outputs(left_run.id)
    right_out = tracker.get_run_outputs(right_run.id)

    left_mode = (
        consist.load_df(left_out["mode_shares"]).sort_values("iteration").iloc[-1]
    )
    right_mode = (
        consist.load_df(right_out["mode_shares"]).sort_values("iteration").iloc[-1]
    )
    left_summ = (
        consist.load_df(left_out["iteration_summaries"])
        .sort_values("iteration")
        .iloc[-1]
    )
    right_summ = (
        consist.load_df(right_out["iteration_summaries"])
        .sort_values("iteration")
        .iloc[-1]
    )
    left_pmt = consist.load_df(left_out["pmt_totals"]).sort_values("iteration").iloc[-1]
    right_pmt = (
        consist.load_df(right_out["pmt_totals"]).sort_values("iteration").iloc[-1]
    )

    rows = []
    for mode in ["car", "transit", "walk"]:
        rows.append(
            {
                "metric": f"mode_share.{mode}",
                "baseline": float(left_mode[mode]),
                "policy": float(right_mode[mode]),
            }
        )
    for metric in ["vmt", "avg_trip_distance"]:
        if metric in left_summ and metric in right_summ:
            rows.append(
                {
                    "metric": f"summary.{metric}",
                    "baseline": float(left_summ[metric]),
                    "policy": float(right_summ[metric]),
                }
            )
    for mode in ["car", "transit", "walk"]:
        rows.append(
            {
                "metric": f"pmt.{mode}",
                "baseline": float(left_pmt[mode]),
                "policy": float(right_pmt[mode]),
            }
        )

    df = pd.DataFrame(rows)
    df["delta"] = df["policy"] - df["baseline"]
    df["pct_delta"] = (df["delta"] / df["baseline"] * 100).round(2)
    return df.sort_values("metric")


outcome_comparison = summary_pair.apply(compare_outcomes)
outcome_comparison
````

## Matrix Comparison Across Runs

Many transportation models produce large skim matrices: travel times, costs, distances, and generalized costs by origin, destination, mode, period, scenario, and iteration. In a regional model those arrays can be much too large to casually read into memory just to answer one comparison question.

`tracker.load_matrix(...)` is meant for that case. It finds the tracked `skims` artifacts from the selected runs, opens the underlying Zarr stores lazily through xarray, and returns one dataset indexed by `run_id` with `year` and `iteration` carried as coordinates. That gives us a scenario-comparison view without manually finding artifact paths, opening multiple stores, or aligning dimensions by hand.

The useful details are:

- **Lazy access**: xarray can defer reading matrix chunks until a slice or calculation actually needs them, which matters for very large skim stores.
- **Variable selection**: `variables=["time_car_mins"]` narrows the view to the measure we want instead of loading every skim variable.
- **Provenance-aware selection**: we can choose runs by explicit `run_ids` here, or use filters such as `parent_id`, `model`, and `status` when we want Consist to find the relevant artifacts from the run database.
- **Consistent coordinates**: once loaded, normal xarray selection syntax lets us compare the same origin/destination cells across scenarios using labeled dimensions instead of positional indexing.

This demo uses a tiny 5-zone matrix so the output is readable, but the same pattern is intended for much larger skim matrices.

````python
baseline_run = tracker.find_latest_run(
    parent_id=SCENARIO_RUN_ID,
    model="traffic_simulation",
    status="completed",
)
parking_run = tracker.find_latest_run(
    parent_id=high_parking_results["scenario_run_id"],
    model="traffic_simulation",
    status="completed",
)

matrix = tracker.load_matrix(
    "skims",
    variables=["time_car_mins"],
    run_ids=[baseline_run.id, parking_run.id],
)

baseline_skims = matrix.sel(run_id=baseline_run.id)
parking_skims = matrix.sel(run_id=parking_run.id)

downtown_zone = 3
baseline_to_cbd = baseline_skims.sel(destination=downtown_zone)["time_car_mins"]
parking_to_cbd = parking_skims.sel(destination=downtown_zone)["time_car_mins"]

compare = pd.DataFrame(
    {
        "origin": baseline_to_cbd.origin.values,
        "baseline_mins": baseline_to_cbd.values,
        "high_parking_mins": parking_to_cbd.values,
    }
)
compare["delta_mins"] = compare["high_parking_mins"] - compare["baseline_mins"]
compare.round(2).sort_values("delta_mins", ascending=False)
````
